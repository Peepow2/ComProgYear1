```python
EL = 43 # ใส่จำนวนอักษรแต่ละบรรทัด
NL = 32 # ใส่จำนวนบรรทัด 
NC = 11 # ใส่จำนวนตัวอักษรในแต่ละบรรทัด

# หากมีคำอยูในใจก็พิมพ์เลย ตามความยาวที่ได้ ถ้าไม่มีเว้นไว้
txt = "" 
```

```python
def generate_txt():
    txt = ""
    temp = "ABCDEFGHIJKLMNOPQRSTUVWXYZ" + "abcdefghijklmnopqrstuvwxyz" + "0123456789"
    temp = list(temp); random.shuffle(temp)
    for c in temp: txt += c
    return txt
```

# **Version มีกรอบ**

```python
import random
def show_txt1(txt, EL, NL, NC):
    EL -= 2; NL -= 2;
    
    if txt == '': txt = generate_txt()
    LINE = set(); LT = len(txt)
    while len(LINE) != NC - 3:
        idx = random.randint(0, LT - 1)
        LINE.add(txt[idx])
        
    # print(LINE)
    LINE = list(LINE); tod = ''
    print('print("' + '+' + '-' * EL + '+'  + '"' + ')')
    for i in range(0, NL):
        tod = ''
        for i in range(0, EL):
            idx = random.randint(0, NC - 4)
            tod += LINE[idx]
        print('print("' + '|' + tod + '|' +  '"' + ")")
    print('print("' + '+' + '-' * EL + '+' + '"' + ")")
```

```python
show_txt1(txt[0: NC], EL, NL, NC)
```

```python
import random
def show_txt2(txt, EL, NL, NC):
    
    if txt == '': txt = generate_txt()
    LINE = set(); LT = len(txt)
    while len(LINE) != NC:
        idx = random.randint(0, LT - 1)
        LINE.add(txt[idx])

    # print(LINE)
    LINE = list(LINE); tod = ''
    for i in range(0, NL):
        tod = ''
        for i in range(0, EL):
            idx = random.randint(0, NC - 1)
            tod += LINE[idx]
        print('print("' + tod  +  '"' + ")")
```
